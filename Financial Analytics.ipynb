```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://drive.google.com/uc?id=1a93WN4_boa__DUIt7l3YrUvfF-Kr4vIi"
data = pd.read_csv(url)
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.histplot(data['Mar Cap - Crore'], bins=20, color='skyblue')
plt.title('Market Capitalization Distribution')
plt.xlabel('Market Cap - Crore')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
sns.histplot(data['Sales Qtr - Crore'], bins=20, color='salmon')
plt.title('Quarterly Sales Distribution')
plt.xlabel('Sales Qtr - Crore')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Mar Cap - Crore', y='Sales Qtr - Crore', data=data, color='coral')
plt.title('Market Capitalization vs. Quarterly Sales')
plt.xlabel('Market Cap - Crore')
plt.ylabel('Sales Qtr - Crore')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data['Mar Cap - Crore'], color='skyblue')
plt.title('Distribution of Market Capitalization')
plt.xlabel('Market Cap - Crore')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data['Sales Qtr - Crore'], color='salmon')
plt.title('Distribution of Quarterly Sales')
plt.xlabel('Sales Qtr - Crore')
plt.show()
```

```python
sns.pairplot(data, vars=['Mar Cap - Crore', 'Sales Qtr - Crore'])
plt.show()
```
